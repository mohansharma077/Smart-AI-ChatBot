```python
#import libraries
from newspaper import Article
import random
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
 #download punkt package
 nltk.download('punkt',quiet=True)
```

```python
#download the url
article=Article('https://en.wikipedia.org/wiki/Artificial_intelligence')
article.download()
article.parse()
article.nlp()
corpus=article.text
print(corpus)
```

```python
#tokenization
text=corpus
sentence_list=nltk.sent_tokenize(text)
```

```python
print(sentence_list)
```

```python
#a function to return random greeting response to a user
```

```python
def greeting_response(text):
  text=text.lower()
  #bots greeting response
  bots_greeting=['hi','hello','hey']
  user_greeting=['greeting','hi','hello','hey']
  for word in text.split():
    if word in user_greeting:
      return random.choice(bots_greeting)

```

```python
def index_sort(list_var):
  length=len(list_var)
  list_index=list(range(0,length))
  x=list_var

  for i in range(length):
    for j in range(length):
      if x[list_index[i]] > x[list_index[j]]:
           #swap
           temp=list_index[i]
           list_index[i]=list_index[j]
           list_index[j]=temp
  return list_index

```

```python
#create the bot response
def bot_response(user_input):
  user_input=user_input.lower()
  sentence_list.append(user_input)
  bot_response=''
  cv=CountVectorizer().fit_transform(sentence_list)
  similarity_score=cosine_similarity(cv[-1],cv)
  similarity_score_list=similarity_score.flatten()
  index = index_sort(similarity_score_list)
  index=index[1:]
  response_flag=0
  j=0
  for i in range(len(index)):
    if similarity_score_list[index[i]]> 0.0:
      bot_response=bot_response + ''+ sentence_list[index[i]]
      response_flag=1
      j=j+1
    if j >2:
      break
  if response_flag==0:
    bot_response =bot_response + ' ' + "I Apolize, I don't understand"
  sentence_list.remove(user_input)
  return bot_response
```

```python

#start the chat
print('Doc BOT: I am Doctor Bot or Doc Bot for short. I will answer your queries about chronic Kidney diseases.If you have any queries??  ')
exit_list=['exit','see you later','bye','quit','break']
while(True):
  user_input=input()
  if user_input.lower() in exit_list:
    print('Doc Bot: Chat with you later!')
    break
  else:
    if greeting_response(user_input)!=None:
      print('Doc Bot:'+greeting_response(user_input))
    else:
      print('Doc Bot:' +bot_response(user_input))

```
